# Spliting data w/ GroupKFold

```python
#group k fold leaves one group out for test
```

```python
from sklearn.model_selection import GroupKFold
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import numpy as np
#leave space for gridsearchcv

random_state = 37
n_folds = 3

#split X and y 
y = df['Rice Yield (kg/ha)']
X = df.drop(columns=['Rice Yield (kg/ha)'])
X, y = shuffle(X, y, random_state=random_state)

#train_test split
X_other, X_test, y_other, y_test = train_test_split(X,y,train_size = 0.8,
                                                    random_state=random_state,shuffle=True)

#groupkfold for validation and training split
groups = X_other['District']
#print(groups)
#print(np.unique(groups))
#print(np.unique(X_test['District']))
kf = GroupKFold(n_splits=3)
for train_index, val_index in kf.split(X_other, y_other, groups=groups): 
    X_train = X_other.iloc[train_index]
    y_train = y_other.iloc[train_index]
    X_val = X_other.iloc[val_index]
    y_val = y_other.iloc[val_index]
    print('training set:',X_train.shape, y_train.shape)
    print('validation set:',X_val.shape, y_val.shape)
    print('test set: ', X_test.shape, y_test.shape)
    
print(np.unique(X_train['District']))
#check reproducibility 
#print(X_other.head(10)) 
#print(X_val.head(1)) 
#print(X_test.head(1))
```

```python
#after train_test split's class balance
plt.figure(figsize=(5,3))
pd.value_counts(X_train['District']).plot.bar()
plt.xticks(rotation=0)
plt.ylabel('count')
plt.xlabel('District')
plt.show()
```

# Preprocessing

Few Ideas might work with geo location features:

- Choosing a model that would not require you to normalize such features
- reverse geocoding (doubt that it will work in this case)
- map these two features to 3 dimnension: x = cos(lat) * cos(lon); y = cos(lat) * sin(lon); z = sin(lat) 

```python
print(X_train['District'].describe())
```

```python
from sklearn.compose import ColumnTransformer 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler, OrdinalEncoder

#categorical feature
cat_ftrs = ['District', 'Season(SA = Summer Autumn, WS = Winter Spring)']
#ordinal feature
ord_ftrs = ['Rice Crop Intensity(D=Double, T=Triple)']
ord_cats = [['D','T']]
#standard scaler
std_ftrs = ['Field size (ha)']

# one-hot encoder
cat_transformer = Pipeline(steps=[('onehot', OneHotEncoder(sparse=False,handle_unknown='ignore'))])

# ordinal encoder
ord_transformer = Pipeline(steps=[('ordinal', OrdinalEncoder(categories = ord_cats))])

# standard scaler
std_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# collect the transformers
preprocessor = ColumnTransformer( 
    transformers=[
        ('onehot', cat_transformer, cat_ftrs),
        ('std', std_transformer, std_ftrs),
        ('ordinal', ord_transformer, ord_ftrs)])

#fit_transform the training set
X_prep = preprocessor.fit_transform(X_train)

#the feature names after fit
feature_names = preprocessor.get_feature_names_out()

#transform the train
df_train= pd.DataFrame(data=X_prep,columns=feature_names)
print('Shape of transformed X_train:',df_train.shape)

#transform the validation
X_val_prep = preprocessor.transform(X_val)
df_val = pd.DataFrame(data=X_val_prep,columns = feature_names)
print('Shape of transformed X_test:',df_val.shape)
print('Feature name:', feature_names)

#transform the test
X_test_prep = preprocessor.transform(X_test)
df_test = pd.DataFrame(data=X_test_prep,columns = feature_names)
print('Shape of transformed X_test:',df_test.shape)
print('Feature name:', feature_names)
```
